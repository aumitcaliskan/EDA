```python
def fill_value(df2):
    if (df2['smoking'] == 'yes') & (df2['bmi'] > 30):
        return 'smoking and obese'
    if (df2['smoking'] == 'yes') & (df2['bmi'] < 30):
        return 'smoking and normal'
    else:
        return 'standard'
```

```python
df2['segments'] = df2.apply(fill_value,axis=1)
```

```python
df2.head()
```

```python
sns.scatterplot(x='age',y='insurance_price', data=df2, hue='segments')
```

```python
df2['bmi_labels']=pd.cut(df2['bmi'],bins=[0,18.5,24.9,29.9,35,99], labels = ['underweight','normal','overweight','obese','ext.obese'])
```
