# Outlier

## Tukey Rule

```python
q1 = df['age'].quantile(0.25)
```

```python
q3 = df['age'].quantile(0.75)
```

```python
iqr = q3-q1
```

```python
ub = (1.5*iqr) + q3
```

```python
lb = q1 - (1.5*iqr)
```

```python
ub
```

```python
lb
```

```python
df['age'][df['age'] > ub]
```
